```python
import feedparser as fp
import json
import newspaper
import pandas as pd
import csv
```

```python
LIMIT = 1000000000
articles_array = []

data = {}
data['newspapers'] = {}

with open('input/journals.json') as data_file:
    companies = json.load(data_file)
```

```python
#scraping newspapers
count = 1
for company, value in companies.items():
        print("Descargando art√≠culos de: ", company)
        paper = newspaper.build(value['link'], memoize_articles=False)
        newsPaper = {
            "link": value['link'],
            "articles": []
        }
        noneTypeCount = 0
        for content in paper.articles:
            if count > LIMIT:
                break
            try:
                content.download()
                content.parse()
            except Exception as e:
                print(e)
                print("continuing...")
                continue

            article = {}
            article['title'] = content.title
            article['text'] = content.text
            article['authors'] = content.authors
            article['top_image'] =  content.top_image
            article['movies'] = content.movies
            article['journal'] = company
            article['published'] = content.publish_date
            article['article_link'] = content.canonical_link
            article['keywords'] = content.keywords
            article['summary'] = content.summary
            article['html'] = content.html
            newsPaper['articles'].append(article)
            articles_array.append(article)
            print(count, "articulos descardos de", company, " using newspaper, url: ", content.url)
            count = count + 1
            #noneTypeCount = 0
        count = 1
        data['newspapers'][company] = newsPaper
```

```python
pd.DataFrame(articles_array).to_csv("output/file.csv")
```
